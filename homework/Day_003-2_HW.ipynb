```python
## 假如我們不想把資料載到自己的電腦裡?
# 把連結填入
target_url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
```

```python
import requests
response = requests.get(target_url)
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[0:100]
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = '\n'

data = data.split(split_tag)
print(len(data))
data[0]
```

```python
data[0]
```

## 將 txt 轉成 pandas dataframe

```python
import pandas as pd
def get_content():
    for line in data:
        part1 = line.split('\t')
        try:
            yield part1[0], part1[1]
        except:
            pass
result = get_content()
label = []
imageUrl = []
result
for p1,p2 in result:
    label.append(p1)
    imageUrl.append(p2)

df = pd.DataFrame([label, imageUrl]).transpose()
df.columns = ['name', 'links']
df
# arrange_data = data
    
# df = pd.DataFrame(arrange_data)
# df.head()
```

```python
def img2arr_fromURLs(url_list, resize = False):
    """
    請完成這個 Function
    Args
        - url_list: list of URLs
        - resize: bool
    Return
        - list of array
    """
    img_list = []
    for url in url_list:
        response = requests.get(url)
        print(url)
        try:
            img = Image.open(BytesIO(response.content))
            if resize:
                img = img.resize((256,256)) # 256 x 256
            img_list.append(img)
        except:
            print('link fail!')
    return img_list
```

```python
result = img2arr_fromURLs(df[0:5]['links'].values, True)
print("Total images that we got: %i " % len(result)) # 如果不等於 5, 代表有些連結失效囉

for im_get in result:
    plt.imshow(im_get)
    plt.show()
```
