```python
# rand：数组元素在0-1之间 (# 均匀分布)
np.random.rand(3, 4)

# randint：指定上下限的整数 (给定上下限范围选取随机整数，生成数组)
np.random.randint(10, 15, (3, 4))

# uniform：指定上下限的小数
np.random.uniform(10, 15, (3, 4))

# randn：标准正态分布 (# 标准正态分布。平均数为0,标准差为1)
np.random.randn(3, 4)

# normal：指定均值、标准差的正态分布
np.random.normal(0, 1, (3, 4))
```

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = np.random.uniform(-1.0,1.0,size=[10, 10])

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.uniform(-1.0,1.0,size=[1000, 3])

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
